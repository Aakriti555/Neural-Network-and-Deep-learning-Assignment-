## Introduction to Deep Learning 


Deep learning is a subset of machine learning inspired by the neural
networks in the human brain. Deep learning is used for  computational advancements and availability of the huge volume of data.

#### What is deep learning?


Deep Learning is ANN(Artifical neural networks) with many layers. 
We build an ANN with several layers where each layer is responsible for learning the intricate patterns in the data. 
Deep ANNs employ several layers, it learns the complex intrinsic features and multi-level abstract representation of data by itself.
In deep learning we use numerous hyperparameters that act as a tuning knob to obtain better results.

### ANN and its layers

Artificial neurons are arranged in layers. Each and every layer will be connected in such a way that information is passed from one layer to another.

A typical ANN consists of the following layers:
- Input layer
- Hidden layer
- Output layer

 

1. Input layer 

    This layer provide input to the network. The number of neurons in the input layer is the number of inputs we feed to the network.

2. Hidden layer

    Any layer between the input layer and the output layer is called a hidden layer. It processes the input received from the input layer. The hidden layer is responsible for deriving complex relationships between input and output. The network is called a deep neural network when we have many hidden layers.

3. Output layer 

    After processing the input, the hidden layer sends its result to the output layer. The number of neurons in the output layer is based on the type of problem we want our network to solve.

        


### Exploring activation functions

Activation Functions are used to introduce non-linerarity in neural networks. 
F(z) , where **z = (input * weights) + bias** and f(.) is the activation function. 
The aim of the activation function is to introduce a non-linear transformation to learn the complex underlying patterns in the data.

#### 1. The sigmoid function
The sigmoid function is one of the most commonly used activation functions.  The sigmoid function is also known as a **logistic** function. It scales the value between 0 and 1 which is used for predicting the probabilty of output.



The sigmoid function can be defined in Python as follows:


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def sigmoid(x):
    return 1/ (1+np.exp(-x))
```

```python
x_data = np.linspace(-20, 20, 1000)
```

```python
data_sigmoid = [sigmoid(x) for x in x_data]
```

```python
plt.plot(x_data, data_sigmoid)
plt.xlabel("X")
plt.ylabel("Y=Sigmoid(X)")
```

#### 2. The tanh function

A hyperbolic tangent (tanh) function outputs the value between -1 to +1 and is expressed as follows: 



The tanh function is implemented as follows:



```python
def tanh(x):
    numerator = 1 - np.exp(-2 * x)
    denominator = 1 + np.exp(-2 * x)
    return numerator / denominator

x_data = np.linspace(-6, 6, 100)

data_tanh = [tanh(x) for x in x_data]

plt.plot(x_data, data_tanh)
plt.xlabel("X")
plt.ylabel("Y=tanh(X)")
```

#### 3. The Rectified Linear Unit function(ReLU)

ReLU outputs a value from 0 to infinity.  It is basically a piecewise function and can be expressed as follows:



It can be expressed as:
F(x) = max(0,x)

when we feed any negative input to the ReLU function, it converts it to zero. The snag for being zero for all negative values is a problem called **dying ReLU**, and a neuron is said to be dead if it always outputs zero.

```python
def ReLU(x):
    if x < 0:
        return 0
    else:
        return x
```

```python
x_data = np.linspace(-15, 15, 100)
```

```python
data_ReLU = [ReLU(x) for x in x_data]
plt.plot(x_data, data_ReLU)
plt.xlabel("X")
plt.ylabel("Y=ReLU(X)")
```

#### 4. The leaky ReLU function

**Leaky ReLU** is a variant of the ReLU function that solves the dying ReLU problem. Instead of converting every negative input to zero, it has a small slope for a negative value.

Leaky ReLU can be expressed as follows:



The value of alpha is typically set to 0.01. we can send them as a parameter to a neural network and make the network learn the optimal value of alpha. 

```python
def leaky_ReLU(x, alpha=0.1):
    if x < 0:
        return alpha * x
    else:
        return x 
    
x_data = np.linspace(-2000, 2000, 1000)
data_leaky_ReLU = [leaky_ReLU(x) for x in x_data]
plt.plot(x_data, data_leaky_ReLU)
plt.xlabel("X")
plt.ylabel("Y=Leaky_ReLU(X)")
```

#### 5. The Exponential linear unit function

Exponential linear unit (ELU), has a small slope for negative values. But instead of having a straight line, it has a log curve.
It can be expressed as follows: 



The ELU function is implemented in Python as follows:

```python
def ELU(x,alpha=1):
    if x < 0:
        return alpha * (np.exp(x) - 1)
    else:
        return x
    
x_data = np.linspace(-10, 10, 100)
data_ELU = [ELU(x) for x in x_data]
plt.plot(x_data, data_ELU)
plt.xlabel("X")
plt.ylabel("Y=ELU(X)")
```

#### 6. The Swish function

Swish Function is a non-monotonic funcation, which means it is neither always non-increasing nor non-decreasing. It provides better performance than ReLU.



Here, sigma(x) is the sigmoid function.



When the value of Beta is 0, then we get the identity function f(x)=x.

It becomes a linear function and, when the value of Beta tends to infinity, then f(x) becomes 2max(0,x), which is basically the ReLU function multiplied by some constant value. So, the value of beta acts as a good interpolation between a linear and a nonlinear function.

The swish function can be implemented as shown:

```python
def swish(x, beta):
    return 2 * x * sigmoid(beta * x) 

x_data = np.linspace(-1000, 1000, 2000)
data_swish = [swish(x, 0.01) for x in x_data]
plt.plot(x_data, data_swish)
plt.xlabel("X")
plt.ylabel("Y=Swish(X)")
```

#### 7. The softmax function

Softmax function is usually applied to the final layer of the network and while performing multi-class classification tasks. It gives the probabilities of each class for being output and thus, the sum of softmax values will always equal 1.
It can be represented as follows:



The softmax function can be implemented in Python as follows:

```python
def softmax(x):
    return np.exp(x) / np.exp(x).sum(axis=0)

x_data = [[1, 2, 3], [2, 3, 4]]
data_softmax = [softmax(x) for x in x_data]
data_softmax
```

The softmax function converts their inputs to probabilities:


```python
print(np.sum(data_softmax[0]))
print(np.sum(data_softmax[1]))
```

### Forward propagation in ANN

The number of layers in a network is equal to the number of hidden layers plus the number of output layers.
Consider a two-layer neural network with one input layer, X, one hidden layer, H, and one output layer, Y, as shown in the following diagram:



Since we have two inputs, the number of neurons in the input layer will be two. We set the number of neurons in the hidden layer to four, and, the number of neurons in the output layer to one. Now, the inputs will be multiplied by weights, and then we add bias and propagate the resultant value to the hidden layer where the activation function will be applied.

The weight and the bias value between the input to the hidden layer are represented by Wxh and Bh, respectively. 

The dimensions of the weight matrix must be number of neurons in the current layer x number of neurons in the next layer.



Here, Z1 =  input*weights+bias 

Let's use the sigmoid x activation function :



We multiply alpha1 by the weight matrix, Why and add bias, By,  and pass the result Z1 to the next layer, which is the output layer:



Now, applying a sigmoid function to Z1, which will result an output value:



This whole process from the input layer to the output layer is known as forward propagation.

The preceding forward propagation steps can be implemented in Python as follows:

```python
def forward_prop(X, Wxh, Why):
    z1 = np.dot(X, Wxh) + bh
    a1 = sigmoid(z1)
    z2 = np.dot(a1, Why) + by
    y_hat = sigmoid(z2)
    return z1, a1, z2, y_hat
```

### How does ANN learn?

Artificial Neural Networks (ANNs) learn by adjusting weights and biases to minimize the error (cost) in predictions. The learning process follows these steps:

Forward Propagation: The network makes predictions using input data and passes it through the layers, calculating an output.

Calculate Loss: The difference between the predicted output and the actual output is measured using a loss (cost) function.

Backpropagation: The error is propagated backward from the output to input layers, calculating gradients (derivatives) of the loss with respect to weights.



Gradient Descent: Using these gradients, the weights are updated by moving in the direction that reduces the error. The update follows the formula:

**weights = weights    gradient**

where  is the learning rate.
This process is repeated over several iterations (epochs) to find optimal weights and minimize the loss.

### Building a Neural Network from scratch

- Here, we will build a neural network that learns the XOR logic.
- It will have two nodes in input layer, a hidden layer with five nodes and an ouput layer with one node.



We will understand step-by-step how a neural network learns the XOR logic:



1. Preparing the data as shown in XOR table:

```python
X = np.array([ [0, 1], [1, 0], [1, 1],[0, 0] ])
y = np.array([ [1], [1], [0], [0]]) 
```

2. Define the number of nodes in each layer:

```python
num_input = 2
num_hidden = 5
num_output = 1
```

3. Initialize the weights and bias randomly. First, we initialize the input to hidden layer weights:

```python
Wxh = np.random.randn(num_input,num_hidden)
bh = np.zeros((1,num_hidden))
```

4. We initialize the hidden to output layer weights:

```python
Why = np.random.randn (num_hidden,num_output)
by = np.zeros((1,num_output))
```

5. Define the sigmoid activation function:


```python
def sigmoid_derivative(z):
    '''
    Returns the value of 1st derivative of the sigmoid function for a value z
    Used by the backpropogation algorithm 
    '''
    return np.exp(-z)/((1+np.exp(-z))**2)
```

6. Define the cost function:

```python
def cost_function(y, y_hat):
    J = 0.5*sum((y-y_hat)**2)
    return J
```

7. Set the learning rate and the number of training iterations:


```python
alpha = 0.01
num_iterations = 5000
```

8. let's start training the network with the following code:

```python
cost =[]
for i in range(num_iterations):
    z1,a1,z2,y_hat = forward_prop(X,Wxh,Why)
    dJ_dWxh, dJ_dWhy = backword_prop(y_hat, z1, a1, z2)
    #update weights
    Wxh = Wxh -alpha * dJ_dWxh
    Why = Why -alpha * dJ_dWhy
    #compute cost
    c = cost_function(y, y_hat)
    cost.append(c)
```

9. Plot the cost function

```python
plt.grid()
plt.plot(range(num_iterations),cost)
plt.title('Cost Function')
plt.xlabel('Training Iterations')
plt.ylabel('Cost')
```
